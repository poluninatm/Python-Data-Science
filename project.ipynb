## Разбиение на тестовый и валидационный сеты, подготовка данных

```python
from sklearn.model_selection import train_test_split
train, valid = train_test_split(data, test_size=0.3, random_state=111)

train.shape, valid.shape
```

```python
distr_info1 = train['DistrictId'].value_counts(normalize=True).reset_index().rename(columns={'index':'DistrictId','DistrictId':'flat_qty_distr'})
```

```python
def add_district_info1(df, distr_info1):
    df = pd.merge(df, distr_info1, on='DistrictId', how='left')
    df['flat_qty_distr'] = df['flat_qty_distr'].fillna(0.000143)
    return df

def add_cat_fts(df, cat_fts=('Ecology_2', 'Ecology_3', 'Shops_2')):
    for col in cat_fts:
        df[col] = (df[col] == 'B').astype(int)
    return df


def prepare_data(df, distr_info1):
    df = add_district_info1(df, distr_info1)
    df = add_cat_fts(df)
    df['Healthcare_1'] = df['Healthcare_1'].fillna(0)
    return df
```

```python
train = prepare_data(train, distr_info1)
valid = prepare_data(valid, distr_info1)
test = prepare_data(test, distr_info1)
valid.head()
```

## Подбор параметров модели и обучение

```python
from sklearn.ensemble import RandomForestRegressor as RF
from sklearn.metrics import r2_score as r2

train.columns
```

```python
# Жадный алгоритм подбора участвующих в модели полей
def search_best_fields():
    model = RF(n_estimators=300, max_depth=18, random_state=42, max_features=1, min_samples_leaf=2)
    all_fields=['flat_qty_distr','Square', 'Floor', 'Shops_2', 'Ecology_1', 'HouseFloor', 'Ecology_2', 'HouseYear','Ecology_3', 'Healthcare_1']
    #'Rooms',  'Shops_1' - параметр был убран, тк опытным путем было установлено, что комбинация без него дает лучший результат
    # 'HouseYear', 'HouseFloor' = Наличие данных в модели улучшало результаты, несмотря на то, что жадных их находил не всегда - зафиксировала
    fields=['HouseYear', 'HouseFloor'] 
    
    max_all=max_=0
    fields_iter=[]

    while(len(fields)<10):

        for field in all_fields:
            if field in fields:
                continue

            fields_tmp=fields[:]
            fields_tmp.append(field)
            model.fit(train.loc[:, fields_tmp], train['Price'])
            pred_valid = model.predict(valid.loc[:, fields_tmp])
            r2_=r2(valid['Price'], pred_valid)
            
            if r2_ > max_:
                max_=r2_
                fields_iter=fields_tmp
        if max_all>=max_ :
            break

        max_all=max_
        fields=fields_iter[:]
        print(max_all,fields)

    return fields_iter

#fields=search_best_fields()

# 0.570967313338911 ['HouseYear', 'HouseFloor', 'Square']
# 0.7004457726889322 ['HouseYear', 'HouseFloor', 'Square', 'flat_qty_distr']
# 0.7233383362472694 ['HouseYear', 'HouseFloor', 'Square', 'flat_qty_distr', 'Ecology_1']
# 0.7266435145040032 ['HouseYear', 'HouseFloor', 'Square', 'flat_qty_distr', 'Ecology_1', 'Healthcare_1']
# 0.7271218058273896 ['HouseYear', 'HouseFloor', 'Square', 'flat_qty_distr', 'Ecology_1', 'Healthcare_1', 'Shops_2']    
    
#fields=['HouseYear', 'HouseFloor', 'Square', 'flat_qty_distr', 'Ecology_1', 'Healthcare_1', 'Shops_2']

```
